```python
# Challenge1
# below for loops will generate the products of all two three-digit numbers
palindromic = []
for three_digit1 in range(100,1000):
    for three_digit2 in range(100,1000):    
        product = three_digit1 * three_digit2
#We know that the product of two three-digit numbers can 5 or 6 digits. Since we are looking for 
#the largest palindromic number we only consider 6 digit numbers
        if len (str(product))== 6:
            digits = []
#in order to compare the elements of each product we convert string into list called "digits"
            for digit in str(product):
                digits.append(digit)
#in order to create palindromic list for a 6-digit number we should have the following logic:
# 1st element == last element AND 2nd element == 5th element AND 3rd element == 4th element
            if digits[5] == digits[0] and digits[4] == digits[1] and digits[3] == digits[2]:
                palindromic.append(product)
#return largest element of palindromic list                            
print ('The largest palindrome made from the product of two three-digit numbers is', max(palindromic))
        
```

```python
# Challenge2
#initialized prime_list with number 2
prime_list=[2]

#create a blank residual_list
residual_list=[]

#generate the list of numbers from 3 to 1999
list_numbers = range(3,2000)

#for number i in list_numbers, if i%j (j is prime_list) is not 0 for all j's in prime_list, then i is a prime number as well
# and i will be added to prime_list
for i in list_numbers:    
    product_of_residuals = 1
    residual = 0
    residual_list=[]
    for j in prime_list:
        residual = i%j  
        residual_list.append(residual)    
    for z in residual_list:
        product_of_residuals = product_of_residuals*z     
    if  product_of_residuals != 0:
        prime_list.append (i)

#calculate the total of prime numbers in prime_list
total = 0
for num in prime_list:
    total = total + num
print (total, 'is the sum of all the primes below 2,000')
```

```python
# Challenge3

#create a blank num_list
num_list=[]

# generate list of numbers from 1 to 999
list_numbers = range (1,1000)

# for all numbers in list of numbers, if it is multiples of 3 or 5, add that number to num_list
for i in list_numbers:
    if i%3 == 0 or i%5 == 0 :
        num_list.append(i)

#calculate the total of numbers in num_list
total = 0
for i in num_list:
    total = total + i
print (total, 'sum of all the multiples of 3 or 5 below 1,000')
```

```python
# Challenge4

#assing any randome string to text
text = 'aAcdD'

#create a blank text_list
text_list=[]

#calculate length of the text variable
text_length = range(0,len(text))

#initialized count variable
count = 1

#create a blank compressed_list
compressed_list=[]

#initialized not_compressable_count variable
not_compressable_count = 0

# convert string into a list with all lowercase elements
for i in text:
    text_list.append(i.lower())

# If the “compressed” string would not become smaller than the original string, return the original string
#if the value of not_compressable_count ==0, it means the string would not become smaller
for z in text_length:
    if text_list[z] == text_list[z-1]:
        not_compressable_count = not_compressable_count + 1

#beloww loop counts repeated letters and concatenates it the count and corresponding letter         
for j in text_length:
    if j == len(text) - 1:         
         compressed_list.append(text_list[j]+ str(count))   
    elif text_list[j] == text_list[j+1]:
        count = count + 1
    elif text_list[j] != text_list[j+1] :
        #print (text_list[j], count)
        compressed_list.append(text_list[j]+ str(count))
        count = 1
        
#convert the element of list into a string
final_word =''
for k in compressed_list:
    final_word = final_word + str(k)
    
#print the final_word, if it can not become smaller, print the original string
if not_compressable_count == 0:
    print (text)
else:
    print(final_word, 'is the compressed word')    
        

    
```
